```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# Load the dataset
df = pd.read_csv("C:\\Users\\AVANI S\\Downloads\\archive (10)\\cleaned.csv")
```

```python
df.head(5)
```

```python
# Check for missing values
print(df.isnull().sum())

# Drop rows with missing values or handle them accordingly
df.dropna(inplace=True)
```

# Road Surface Type and Accident Severity

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualize accident severity by road surface type
sns.countplot(x='Road_surface_type', hue='Accident_severity', data=df)
plt.title('Accident Severity by Road Surface Type')
plt.xticks(rotation=45)
plt.show()
```

# Weather Conditions and Accident Severity

```python
# Visualize accident severity by weather conditions
sns.countplot(x='Weather_conditions', hue='Accident_severity', data=df)
plt.title('Accident Severity by Weather Conditions')
plt.xticks(rotation=45)
plt.show()
```

# Accident Severity by Types of Junction

```python
# Accident severity by junction type
sns.countplot(x='Types_of_Junction', hue='Accident_severity', data=df)
plt.title('Accident Severity by Types of Junction')
plt.xticks(rotation=45)
plt.show()
```

# Accident Severity by Vehicle Movement

```python
# Accident severity by vehicle movement
sns.countplot(x='Vehicle_movement', hue='Accident_severity', data=df)
plt.title('Accident Severity by Vehicle Movement')
plt.xticks(rotation=45)
plt.show()
```
